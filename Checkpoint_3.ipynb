# **DATA PREPARATION**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/daily_food_nutrition_dataset.csv")

df
```

**Pie Chart**

```python
plt.figure(figsize=(6, 6))
plt.pie(df['Meal_Type'].value_counts(), labels=df['Meal_Type'].value_counts().index, autopct='%1.1f%%', colors=['lightcoral', 'lightblue', 'lightgreen', 'gold'])
plt.title('Tipe Makanan')
plt.show()
```

# Histogram

```python
# Histogram (Matplotlib): Distribusi Durasi Kunjungan
plt.figure(figsize=(8, 5))
plt.hist(df['Category'], bins=20, color='pink', edgecolor='black')
plt.title('Distribusi Kategori Makanan')
plt.xlabel('Kategori Makanan')
plt.ylabel('Jumlah')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='Category', y='Calories (kcal)', hue='Category', data=df, palette='pastel')
plt.title('Distribusi Kalori Berdasarkan Kategori Makanan')
plt.xlabel('Kategori Makanan')
plt.ylabel('Kalori (kcal)')
plt.xticks(rotation=45)
plt.show()
```
